```python
users = pd.read_csv(url, delimiter = '|')
users.head
```

```python
# Step 4. Discover what is the mean age per occupation
```

```python
users[['occupation','age']].groupby('occupation').mean()
```

```python
# Step 5. Discover the Male ratio per occupation and sort it from the most to the least
```

```python
df = users[['gender','occupation']]
Men = df.loc[df['gender'] == 'M']
Total = df.groupby('occupation').count()
Men = Men.groupby('occupation').count()
Diffrence_Ratio = Men / Total
Diffrence_Ratio.sort_values(by='gender', ascending=False)
```

```python
# Step 6. For each occupation, calculate the minimum and maximum ages
```

```python
print('Minimum age ')
df = users[['age','occupation']]
df.groupby('occupation').min()
```

```python
print('Maximum age ')
df.groupby('occupation').max()
```

```python
#Step 7. For each combination of occupation and sex, calculate the mean age
```

```python
Mean_Age = users.groupby(['occupation','gender']).agg({'age':'mean'})
print(Mean_Age)
```

```python
#Step 8. For each occupation present the percentage of women and men
```

```python
Percentage = (users.groupby(['occupation', 'gender'])['gender'].count().groupby(level = 0).transform(lambda x: x/x.sum()*100))
print(Percentage)
```

```python
#Step 1. Import the necessary libraries
```

```python
import pandas as pd
import numpy as np
import datetime
```

```python
#Step 2. Import the dataset from the attached file wind.txt
```

```python
df = pd.read_csv('wind.txt', delimiter='\s+')
df
```

```python
# Step 3. Assign it to a variable called data and replace the first 3 columns by a proper datetime index
```

```python
df.rename(columns={'Yr':'Year','Mo':"Month","Dy":"Day"}, inplace=True)

df.head()
```

```python
#Step 4. Year 2061? Do we really have data from this year? Create a function to fix it and apply it.
```

```python
df['Year'] += 1900
df
```

```python
#Step 5. Set the right dates as the index. Pay attention at the data type, it should be datetime64[ns].
```

```python
df.index = pd.to_datetime(pd.concat([df['Year'], df['Month'], df['Day']], axis=1))
df.drop(['Year','Month','Day'], axis=1, inplace=True)
df
```

```python
#Step 6. Compute how many values are missing for each location over the entire record.They should be ignored in all calculations below.
```

```python
df.isnull().sum()
```

```python
#Step 7. Compute how many non-missing values there are in total.
```

```python
df.notnull().sum()
```

```python
#Step 8. Calculate the mean windspeeds of the windspeeds over all the locations and all the times.A single number for the entire dataset.
```

```python
df.mean().mean()
```

```python
#Step 9. Create a DataFrame called loc_stats and calculate the min, max and mean 
#windspeeds and standard deviations of the windspeeds at each location over all the 
#days
```

```python
Stats = pd.concat([df.mean(), df.min(), df.max(), df.std()], axis=1)
Stats.rename(columns={0:'Mean',1:'Min',2:'Max',3:'Std'}, inplace=True)
Stats
```

```python
#Step 10. Create a DataFrame called day_stats and calculate the min, max and mean 
#windspeed and standard deviations of the windspeeds across all the locations at each 
#day.
```

```python
Dstats = pd.concat([df.mean(axis=1), df.min(axis=1), df.max(axis=1), df.std(axis=1)], axis=1)
Dstats.rename(columns={0:'Mean',1:'Min',2:'Max',3:'Std'}, inplace=True)
Dstats
```

```python
#Step 11. Find the average windspeed in January for each location.Treat January 1961 and January 1962 both as January.
```

```python
January = df[df.index.month == 1]
January.mean()
```

```python
#Step 12. Downsample the record to a yearly frequency for each location.
```

```python
Year = df.resample('Y').ffill()
Year
```

```python
#Step 13. Downsample the record to a monthly frequency for each location.
```

```python
Month = df.resample('M').ffill()
Month
```

```python
#Step 14. Downsample the record to a weekly frequency for each location
```

```python
Week = df.resample('W').ffill()
Week
```

```python
#Step 15. Calculate the min, max and mean windspeeds and standard deviations of the 
#windspeeds across all locations for each week (assume that the first week starts on 
#January 2 1961) for the first 52 weeks.
```

```python
W_stats = Week[:52]
W_stats = pd.concat([W_stats.mean(), W_stats.max(), W_stats.mean(), W_stats.std()], axis=1)
W_stats.rename(columns={0:'Mean',1:'Max',2:'Mean',3:'Std'},inplace=True)
W_stats
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('us-marriages-divorces-1867-2014.csv')
df
```

```python
plt.plot(df['Year'],df['Marriages_per_1000'],label = "Marriages per capita")
plt.xlabel("Year") 
plt.ylabel("Status") 
plt.plot(df['Year'],df['Divorces_per_1000'],'-', label = "Divorces per capita")
plt.legend(loc="upper left")
plt.title("Marriages &. Divorces Rate") 
plt.rcParams ["figure.figsize"] = (15,10)
plt.show()
```

# Question 7

```python
a = df[df['Year']== 1900]
b = df[df['Year']== 1950]
c= df[df['Year']== 2000]
plt.bar(a['Year'], a['Marriages_per_1000'], label = 'Marriages in 1900')
plt.bar(a['Year'], a['Divorces_per_1000'], label = 'Divorces in 1900')
plt.bar(b['Year'], b['Marriages_per_1000'], label = 'Marriages in 1950')
plt.bar(b['Year'], b['Divorces_per_1000'],  label = 'Divorces in 1950')
plt.bar(c['Year'], c['Marriages_per_1000'], label = 'Marriages in 2000')
plt.bar(c['Year'], c['Divorces_per_1000'],  label = 'Divorces  in 2000')
plt.xlabel("Year") # setting x-label as Year
plt.ylabel("Count") # setting y_label as Marriages 
plt.legend(loc="upper left")
plt.title(" Vertical bar graph for marriages and divorces in US per capita for the years 1900,1950 and 2000 respectively ")
#plt.rcParams ["figure.figsize"] = (15,10)
plt.show()
```
